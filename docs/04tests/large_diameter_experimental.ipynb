# Example large diameter well
## Experimental and for radial flow only

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import k0

import timml as tml
```

```python
rw = 5
ml = tml.Model3D(kaq=10, z=np.arange(20, -1, -2), kzoverkh=50)
w = tml.well.LargeDiameterWell(ml, Qw=100, layers=[0, 1, 2, 3, 4], rw=rw)
rf = tml.Constant(ml, 100, 0, 20)
ml.solve()
```

```python
xg = np.linspace(rw, 50, 100)
h = ml.headalongline(xg, 0)
for i in range(10):
    plt.plot(xg, h[i])
plt.xlim(0, 50)
plt.xticks(np.arange(0, 51, 5))
plt.grid()
```

```python
qx = ml.disvec(rw, 0)[0]
print(qx * 2 * np.pi * rw)
print(np.sum(qx * 2 * np.pi * rw))
```
