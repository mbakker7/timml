```python
import matplotlib.pyplot as plt
import numpy as np

import timml as tml
```

```python
ml = tml.ModelMaq(kaq=[10, 20, 5], z=[21, 20, 12, 10, 2, 0, -10], c=[100, 100, 200], 
                  hstar=10, topboundary='semi')
w = tml.Well(model=ml, xw=20, yw=40, rw=0.3, Qw=100, layers=[0, 1])
ml.solve()
print('head inside: ', w.headinside())
print('discharge: ', w.discharge())
```

```python
ml = tml.ModelMaq(kaq=[10, 20, 5], z=[21, 20, 12, 10, 2, 0, -10], c=[100, 100, 200], 
                  hstar=10, topboundary='semi')
w = tml.Well(model=ml, xw=20, yw=40, rw=0.3, Qw=100, res=0.1, layers=[0, 1])
ml.solve()
print('head inside: ', w.headinside())
print('discharge: ', w.discharge())
```

```python
ml = tml.ModelMaq(kaq=[10, 20, 5], z=[21, 20, 12, 10, 5, 0, -10], c=[100, 100, 200], 
                  hstar=10, topboundary='semi')
w = tml.HeadWell(model=ml, xw=20, yw=40, rw=0.3, hw=6, layers=[0, 1])
ml.solve()
print('head inside: ', w.headinside())
print('discharge: ', w.discharge())
```

```python
ml = tml.ModelMaq(kaq=[10, 20, 5], z=[21, 20, 12, 10, 5, 0, -10], c=[100, 100, 200], 
                  hstar=10, topboundary='semi')
w = tml.HeadWell2(model=ml, xw=20, yw=40, rw=0.3, hw=6, res=0.1, layers=[0, 1])
ml.solve()
print('head inside: ', w.headinside())
print('discharge: ', w.discharge())
```
