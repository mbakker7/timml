```python
import timml as tml
import numpy as np
import matplotlib.pyplot as plt
```

```python
ml = tml.ModelMaq(kaq=[1, 2], z=[10, 5, 4, 0], c=20)
xy = [(-5, 0), (0, 0), (5, 0), (5, 8), (-5, 8)]
p1 = tml.PolygonInhomMaq(ml, xy=xy, kaq=[0.2, 8], z=[11, 10, 5, 4, 0], 
                         c=[2, 20], topboundary='semi', hstar=1.0, order=3, ndeg=1)
w = tml.Well(ml, xw=0, yw=-10, Qw=100, layers=1)
rf = tml.Constant(ml, xr=0, yr=-100, hr=2)
ml.solve()

ml.contour([-10, 10, -15, 10], 50, [0, 1], 30)
```
